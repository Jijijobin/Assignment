```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
independ=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independ
```

```python
depend=dataset[['charges']]
depend
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independ,depend,test_size=0.30,random_state=0)
```

```python
pip install xgboost
```

```python
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
import xgboost as xgb
```

```python
regressor= xgb.XGBRegressor(objective ='reg:squarederror', n_estimators=100,  learning_rate=0.1,  max_depth=3,  random_state=42)
regressor=regressor.fit(X_train, Y_train)
```

```python
Y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
r_score
```

```python
import pickle
filename="final_model_xgboost.sav"
```
